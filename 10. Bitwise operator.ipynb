```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
blank = np.zeros((400,400), dtype = 'uint8')
# rectangle(img, pt1, pt2, color[, thickness[, lineType[, shift]]])
# circle(img, center, radius, color[, thickness[, lineType[, shift]]])

rectangle = cv2.rectangle(blank.copy(), (30,30), (370,370), 255, -1)          # here 255 represents white color
circle = cv2.circle(blank.copy(), (200,200), 200, 255, -1)
cv2.imshow('Rectangle', rectangle)
cv2.imshow('Circle', circle)
cv2.waitKey(0)
plt.imshow(rectangle)
plt.imshow(circle)
```

1. Bitwise operator AND  ------> Intersecting region

```python
bitwise_and = cv2.bitwise_and(rectangle, circle)
cv2.imshow('Bitwise And', bitwise_and)
cv2.waitKey(0)
plt.imshow(bitwise_and)
```

2. Bitwise OR ------> Non intersecting and intersecting region

```python
bitwise_or = cv2.bitwise_or(rectangle, circle)
cv2.imshow('Bitwise Or', bitwise_or)
cv2.waitKey(0)
plt.imshow(bitwise_or)
```

3. Bitwise XOR -----> Non intersecting regions

```python
bitwise_xor = cv2.bitwise_xor(rectangle, circle)
cv2.imshow('Bitwise XOR', bitwise_xor)
cv2.waitKey(0)
plt.imshow(bitwise_xor)
```

4. Bitwise NOT

```python
bitwise_not = cv2.bitwise_not(rectangle)
cv2.imshow('Bitwise Not', bitwise_not)
cv2.waitKey(0)
plt.imshow(bitwise_not)
```

```python
bitwise_not = cv2.bitwise_not(circle)
cv2.imshow('Bitwise Not', bitwise_not)
cv2.waitKey(0)
plt.imshow(bitwise_not)
```
