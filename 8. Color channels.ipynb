```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('G:/New folder/harry.jpg')
cv2.imshow('Harry', img)

# Now we will split coloured image (BGR) into B,G,R by cv2.split()

b,g,r = cv2.split(img)

cv2.imshow('Blue', b)
cv2.imshow('Green', g)
cv2.imshow('Red', r)
plt.imshow(b)
# To show the shape of images we will pass

print(img.shape)
print(b.shape)
print(g.shape)
print(r.shape)

cv2.waitKey(0)
```

```python
# Now merge thses images

merged = cv2.merge([b,g,r])        #---passing lists b, g, r
cv2.imshow("Merged", merged)
cv2.waitKey(0)
plt.imshow(merged)
```

```python
# create blank image to split image into blue, green, red colour

blank = np.zeros(img.shape[:2], dtype = 'uint8')

# to print image with blue channel  ,# blank image consists of height and width, not necessarily number of color channel
blue = cv2.merge([b, blank, blank])  # only displays blue channel, setting G & R to blank
green = cv2.merge([blank, g, blank]) # only displays green channel, setting B & R to blank
red = cv2.merge([blank, blank, r])  # only displays red channel, setting B & G to blank

cv2.imshow('Blue', blue)
cv2.imshow('Green', green)
cv2.imshow('Red', red)

cv2.waitKey(0)

plt.imshow(blue)
plt.imshow(green)
plt.imshow(red)
```
