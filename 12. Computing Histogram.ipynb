```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('G:/New folder/Force.jpg')
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow('Gray',gray)
cv2.waitKey(0)
plt.imshow(gray)
```

# Gray scale histogram

```python
# calcHist(images, channels, mask, histSize, ranges[, hist[, accumulate]])
#  Here image is list so we will pass it into [ ], Histsize ----> no. of bins
gray_hist = cv2.calcHist([gray], [0], None, [256], [0,256])       
plt.figure()
plt.title('Gray scale histogram')
plt.xlabel('Bins')
plt.ylabel('# of pixels')
plt.plot(gray_hist)

# To give limit across x axis
plt.xlim([0,256])
plt.show()
```

# Color histogram

```python
img2 = cv2.imread('G:/New folder/animal.jpg')
colors = ('b', 'g', 'r')                      # tuples
plt.figure()
plt.title('Color histogram')
plt.xlabel('Bin')
plt.ylabel('# of pixel')
for i, col in enumerate(colors):
    hist = cv2.calcHist([img2], [i], None, [256], [0,256])
    plt.plot(hist, color = col)
    plt.xlim([0,256])
cv2.waitKey(0)
plt.show()
```
