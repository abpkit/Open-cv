```python
import cv2
import os
import matplotlib.pyplot as plt
face_cascade = cv2.CascadeClassifier('C:/Users/Abhi/Desktop/CSV files/DeepLearningDemo-master/DeepLearningDemo-master/OpenCV/haarcascade_frontalface_default.xml')
```

```python
img = cv2.imread('G:/New folder/harry.jpg')
```

```python
faces = face_cascade.detectMultiScale(img)
print(faces)
```

```python
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (255,0,0), 2)


cv2.imshow('img', img)
plt.imshow(img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
