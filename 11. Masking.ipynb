```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('G:/New folder/harry.jpg')
cv2.imshow('Harry', img)
blank = np.zeros(img.shape[:2], dtype = 'uint8')
cv2.imshow('Blank Image', blank)
cv2.waitKey(0)
plt.imshow(img)
```

```python
mask = cv2.circle(blank, (img.shape[1]//2, img.shape[0]//2), 350, 255, -1)
cv2.imshow('Mask', mask)
plt.imshow(mask)
cv2.waitKey(0)
```

```python
# bitwise_and(src1, src2[, dst[, mask]])
masked = cv2.bitwise_and(img,img, mask=mask)
cv2.imshow('Masked', masked)
cv2.waitKey(0)
plt.imshow(masked)
```
