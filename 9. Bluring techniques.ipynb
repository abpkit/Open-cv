```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

We generally smoothen the image when there is lot of noise in the image. Noise may be due to camera sensors, lighting.
We can smoothen the image by blurring the image

# Averaging method

```python
img = cv2.imread('G:/New folder/Force.jpg')
cv2.imshow('Dada', img)
cv2.waitKey(0)

# By averaging method ---> blur(src, ksize[, dst[, anchor[, borderType]]]) 
# It gives average of surrounding pixels at centre
average = cv2.blur(img, (7,7))
cv2.imshow('Average blur', average)
cv2.waitKey(0)
plt.imshow(average)
```

# Gaussian blur method

```python
# Gaussian blur method ----> we get less blur in this method as compare to averaging method, but it is more natural
# Weight value added in gaussian blur method
#         GaussianBlur(src, ksize, sigmaX[, dst[, sigmaY[, borderType]]])

gauss = cv2.GaussianBlur(img, (7,7), 0)      # here we passed standard deviation as 0.
cv2.imshow('Gaussian Blur', gauss)
cv2.waitKey(0)
plt.imshow(gauss)
```

# Median Blur

 It gives average of surrounding pixels at centre
 It is more effective than averaging and Gaussian blur method to reduce noise
 medianBlur(src, ksize[, dst])

```python
median = cv2.medianBlur(img, 3)            # openCV assumes automatically ksize =3 is ----> 3x3
cv2.imshow('Median Blur', median)
cv2.waitKey(0)
```

# Bilateral blur

```python
bilateral = cv2.bilateralFilter(img, 10, 35, 25)
cv2.imshow('Bilateral Blur', bilateral)
cv2.waitKey(0)
plt.imshow(bilateral)
```
