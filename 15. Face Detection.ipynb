```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python

img = cv2.imread('G:/New folder/baby.png')
cv2.imshow('BABY', img)
cv2.waitKey(0)  
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow('Gray', gray)
cv2.waitKey(0)
plt.imshow(gray)
```

```python
# Reading Haar file

haar_cascade = cv2.CascadeClassifier('C:/Users/Abhi/Desktop/CSV files/DeepLearningDemo-master/DeepLearningDemo-master/OpenCV/haarcascade_frontalface_default.xml')

# detectMultiScale(image[, scaleFactor[, minNeighbors[, flags[, minSize[, maxSize]]]]])
face_rect = haar_cascade.detectMultiScale(gray, scaleFactor = 1.1, minNeighbors = 3)

print(f'Number of faces found = {len(face_rect)}')
```

```python
# How to draw rectangle on face detected

for x,y,w,h in face_rect:
    cv2.rectangle(img, (x,y), (x+w, y+h), (0,255,0), thickness = 11)

cv2.imshow('Detected faces', img)
cv2.waitKey(0)
plt.imshow(img)
```
