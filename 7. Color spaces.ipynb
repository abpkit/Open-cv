```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('G:/New folder/harry.jpg')
cv2.imshow('Harry', img)

# BGR to GRAY scale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow('Harry_gray', gray)
cv2.waitKey(0)
plt.imshow(gray)
```

# BGR to HSV (Huge Saturation Value)

```python
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
cv2.imshow('HSV', hsv)
cv2.waitKey(0)
plt.imshow(hsv)
```

# BGR to LAB  (L*A*B)

```python
lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
cv2.imshow('LAB', lab)
cv2.waitKey(0)
plt.imshow(lab)
```

```python
# OpenCV plots image by BGR and MATPLOTLIB plots image by RGB
# so we see the inversion of colour in both different images

image = cv2.imread('G:/New folder/animal.jpg')
cv2.imshow('Animal', image)
cv2.waitKey(0)
plt.imshow(image)
plt.show()
```

# BGR to RGB

```python
rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
cv2.imshow('RGB', rgb)
cv2.waitKey(0)
plt.imshow(rgb)
```

```python
#HSV to BGR
hsv_bgr = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)
cv2.imshow('HSV-->BGR', hsv_bgr)
cv2.waitKey(0)
plt.imshow(hsv_bgr)
```

```python
# LAB to BGR
lab_bgr = cv2.cvtColor(lab, cv2.COLOR_LAB2BGR)
cv2.imshow('LAB-->BGR', lab_bgr)
cv2.waitKey(0)
plt.imshow(lab_bgr)
```
